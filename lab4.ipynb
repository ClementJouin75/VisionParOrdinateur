#### Reconstruire une vidéo 
nous avons reconstruit une vidéo à partir des images cela peut-etre un peu long avant d'avoir la fenêtre environs 1m30

```python
import cv2
import sys
import numpy as np

ESC_KEY = 27
Q_KEY = 113
M = 200  # Nombre d'images pour calculer la moyenne

def isolate_cars(img_vector_car, road_mask):
    isolated_cars_vector = []
    for img in img_vector_car:
        _, thresholded_frame = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY)
        isolated_cars = cv2.bitwise_and(thresholded_frame, road_mask)
        isolated_cars_vector.append(isolated_cars)
    return isolated_cars_vector

def count_cars(isolated_image):
    # Trouver les contours dans l'image isolée
    contours, _ = cv2.findContours(isolated_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Initialiser le compteur de voitures
    car_count = 0

    # Parcourir tous les contours et compter les voitures
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 50:  # Filtre pour exclure les petits contours parasites
            car_count += 1

    return car_count

def save_images_as_video(images, video_name, fps=30):
    if not images:
        # Si la liste est vide, utilisez une taille par défaut
        height, width = 480, 640
    else:
        height, width = images[0].shape

    fourcc = cv2.VideoWriter_fourcc(*'MJPG')
    out = cv2.VideoWriter(video_name, fourcc, fps, (width, height), isColor=False)

    for image in images:
        out.write(image)

    out.release()

def main():
    # Chemin vers la vidéo
    video_name = "./Fichiers_TP4/Fichiers_autoroute/video.avi"
    print("Nom de la vidéo:", video_name)

    # Ouverture de la capture vidéo
    cap = cv2.VideoCapture(video_name)
    if not cap.isOpened():
        print("L'ouverture de la capture a échoué; nous ne pouvons rien faire :'(")
        return 1

    # Création d'une liste pour stocker les images en niveaux de gris
    img_vector_car = []

    # Création d'une liste pour stocker les images isolées
    isolated_images_vector = []

    # Boucle principale pour lire chaque frame de la vidéo
    frame_number = 0
    while True:
        # Lecture du frame suivant
        ret, im = cap.read()

        if not ret:
            # Impossible de lire le frame suivant, sortie de la boucle
            break

        # Convertir l'image en niveaux de gris
        im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

        # Ajouter l'image en niveaux de gris à la liste
        img_vector_car.append(im_gray)

        # Si le nombre total d'images atteint M, isoler les voitures
        if len(img_vector_car) >= M:
            avg_image = cv2.reduce(np.array(img_vector_car[-M:], dtype=np.uint8), 0, cv2.REDUCE_AVG)[0]
            
            _, thresholded_image_low = cv2.threshold(avg_image, 90, 255, cv2.THRESH_BINARY)
            _, thresholded_image_high = cv2.threshold(avg_image, 195, 255, cv2.THRESH_BINARY)
            
            subtracted_image = cv2.absdiff(thresholded_image_low, thresholded_image_high)
            road_mask = cv2.erode(cv2.dilate(subtracted_image, None, iterations=1), None, iterations=9)

            # Isoler les voitures dans l'image courante en utilisant le masque de route
            isolated_cars_vector = isolate_cars(img_vector_car[-M:], road_mask)

            # Compter et afficher le nombre de voitures
            for idx, isolated_cars in enumerate(isolated_cars_vector):
                car_count = count_cars(isolated_cars)
                print(f"Frame {frame_number - M + idx} - Nombre de voitures: {car_count}")

            # Ajouter les images isolées à la liste
            isolated_images_vector.extend(isolated_cars_vector)

        key = cv2.waitKey(1) & 0xFF
        if key == ESC_KEY or key == Q_KEY:
            break

        frame_number += 1

    cap.release()
    cv2.destroyAllWindows()

    # Enregistrer les images isolées en tant que vidéo
    save_images_as_video(isolated_images_vector, 'output_video.avi')

    # Ouvrir la vidéo générée
    generated_video = cv2.VideoCapture('output_video.avi')
    while True:
        ret, frame = generated_video.read()
        if not ret:
            break

        cv2.imshow('Generated Video', frame)
        if cv2.waitKey(30) & 0xFF == 27:
            break

    generated_video.release()
    cv2.destroyAllWindows()

    return 0

if __name__ == "__main__":
    sys.exit(main())
```

```python
import cv2
import sys
import numpy as np

ESC_KEY = 27
Q_KEY = 113
M = 200  # Nombre d'images pour calculer la moyenne

def isolate_cars(img_vector_car, road_mask):
    isolated_cars_vector = []
    for img in img_vector_car:
        _, thresholded_frame = cv2.threshold(img, 80, 255, cv2.THRESH_BINARY)
        isolated_cars = cv2.bitwise_and(thresholded_frame, road_mask)
        isolated_cars_vector.append(isolated_cars)
    return isolated_cars_vector

def count_cars(isolated_image):
    # Trouver les contours dans l'image isolée
    contours, _ = cv2.findContours(isolated_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Initialiser le compteur de voitures
    car_count = 0

    # Parcourir tous les contours et compter les voitures
    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 50:  # Filtre pour exclure les petits contours parasites
            car_count += 1

    return car_count

def save_images_as_video(images, video_name, fps=30):
    if not images:
        # Si la liste est vide, utilisez une taille par défaut
        height, width = 480, 640
    else:
        height, width = images[0].shape

    fourcc = cv2.VideoWriter_fourcc(*'MJPG')
    out = cv2.VideoWriter(video_name, fourcc, fps, (width, height), isColor=False)

    for image in images:
        out.write(image)

    out.release()

def main():
    # Chemin vers la vidéo
    video_name = "./Fichiers_TP4/Fichiers_autoroute/video.avi"
    print("Nom de la vidéo:", video_name)

    # Ouverture de la capture vidéo
    cap = cv2.VideoCapture(video_name)
    if not cap.isOpened():
        print("L'ouverture de la capture a échoué; nous ne pouvons rien faire :'(")
        return 1

    # Création d'une liste pour stocker les images en niveaux de gris
    img_vector_car = []

    # Création d'une liste pour stocker les images isolées
    isolated_images_vector = []

    # Boucle principale pour lire chaque frame de la vidéo
    frame_number = 0
    while True:
        # Lecture du frame suivant
        ret, im = cap.read()

        if not ret:
            # Impossible de lire le frame suivant, sortie de la boucle
            break

        # Convertir l'image en niveaux de gris
        im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

        # Ajouter l'image en niveaux de gris à la liste
        img_vector_car.append(im_gray)

        # Si le nombre total d'images atteint M, isoler les voitures
        if len(img_vector_car) >= M:
            avg_image = cv2.reduce(np.array(img_vector_car[-M:], dtype=np.uint8), 0, cv2.REDUCE_AVG)[0]
            
            _, thresholded_image_low = cv2.threshold(avg_image, 90, 255, cv2.THRESH_BINARY)
            _, thresholded_image_high = cv2.threshold(avg_image, 195, 255, cv2.THRESH_BINARY)
            
            subtracted_image = cv2.absdiff(thresholded_image_low, thresholded_image_high)
            road_mask = cv2.erode(cv2.dilate(subtracted_image, None, iterations=1), None, iterations=9)

            # Isoler les voitures dans l'image courante en utilisant le masque de route
            isolated_cars_vector = isolate_cars(img_vector_car[-M:], road_mask)

            # Compter et afficher le nombre de voitures
            for idx, isolated_cars in enumerate(isolated_cars_vector):
                car_count = count_cars(isolated_cars)
                print(f"Frame {frame_number - M + idx} - Nombre de voitures: {car_count}")

            # Ajouter les images isolées à la liste
            isolated_images_vector.extend(isolated_cars_vector)

            # Sélectionner une image isolée pour affichage
            selected_frame_index = 0  # Choisissez l'index de la frame que vous souhaitez visualiser
            selected_frame = isolated_images_vector[selected_frame_index]

            # Trouver les contours dans l'image isolée
            contours, _ = cv2.findContours(selected_frame, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            # Initialiser le compteur de voitures
            car_count_selected = 0

            # Parcourir tous les contours et compter les voitures
            for contour in contours:
                area = cv2.contourArea(contour)
                if area > 50:  # Filtre pour exclure les petits contours parasites
                    car_count_selected += 1

            # Dessiner les contours sur l'image pour visualisation
            contours_image = cv2.cvtColor(selected_frame, cv2.COLOR_GRAY2BGR)
            cv2.drawContours(contours_image, contours, -1, (0, 255, 0), 2)

            # Afficher le nombre de voitures
            cv2.putText(contours_image, f"Nombre de voitures: {car_count_selected}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

            # Afficher l'image résultante
            cv2.imshow("Résultat avec le nombre de voitures", contours_image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

        key = cv2.waitKey(1) & 0xFF
        if key == ESC_KEY or key == Q_KEY:
            break

        frame_number += 1

    cap.release()
    cv2.destroyAllWindows()

    # Enregistrer les images isolées en tant que vidéo
    save_images_as_video(isolated_images_vector, 'output_video.avi')

    # Ouvrir la vidéo générée
    generated_video = cv2.VideoCapture('output_video.avi')
    while True:
        ret, frame = generated_video.read()
        if not ret:
            break

        cv2.imshow('Generated Video', frame)
        if cv2.waitKey(30) & 0xFF == 27:
            break

    generated_video.release()
    cv2.destroyAllWindows()

    return 0

if __name__ == "__main__":
    sys.exit(main())
```
